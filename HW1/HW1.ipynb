## 0. Setup

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd

%matplotlib inline
```

## 1. Load Audio

Load the provided audio file (or any `.wav` file of your choice for testing).

```python
AUDIO_PATH = "sample.wav"  # TODO: update path if needed

y, sr = librosa.load(AUDIO_PATH, sr=None)
print(f"Sample rate : {sr} Hz")
print(f"Duration    : {len(y)/sr:.2f} s")
ipd.Audio(y, rate=sr)
```

## 2. Waveform Visualisation

```python
plt.figure(figsize=(12, 3))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

## 3. Spectrogram

```python
# TODO: compute the Short-Time Fourier Transform (STFT) and plot the spectrogram
# Hint: librosa.stft, librosa.amplitude_to_db, librosa.display.specshow

D = librosa.stft(y)
D_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

plt.figure(figsize=(12, 4))
librosa.display.specshow(D_db, sr=sr, x_axis="time", y_axis="hz")
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram")
plt.tight_layout()
plt.show()
```

## 4. Mel-Frequency Cepstral Coefficients (MFCCs)

```python
# TODO: extract MFCCs and plot them
N_MFCC = 13

mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=N_MFCC)

plt.figure(figsize=(12, 4))
librosa.display.specshow(mfccs, sr=sr, x_axis="time")
plt.colorbar()
plt.title(f"MFCCs (n={N_MFCC})")
plt.tight_layout()
plt.show()
```
